```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Criar diretrio de sada se no existir
output_dir = "result"
os.makedirs(output_dir, exist_ok=True)

# Parmetros iniciais
jmax = 500   # nmero de pontos de grade
nmax = 3800  # nmero de passos de tempo

# Condio inicial
fatu = np.zeros(jmax)
fatu[224:275] = 99  # Definir intervalo de j para sinal retangular (ajuste de ndice)
fcin = fatu.copy()

d = 10     # Difusividade em m/s
dt = 2     # Passo de tempo em s
dx = 8     # Espaamento da grade em m
q = d * dt / (dx * dx)
xgrid = np.arange(jmax) * dx

# Esquema explcito avanado no tempo e centrado no espao
fren = fatu.copy()
freqplot = 360
kplot = 359

# Loop temporal
for n in range(2, nmax + 1):
    tempo = n * dt
    kplot += 1
    
    # Frmula explcita para a equao de difuso
    fren[1:jmax-1] = fatu[1:jmax-1] + q * (fatu[2:jmax] - 2 * fatu[1:jmax-1] + fatu[0:jmax-2])
    fren[fren < 0] = 0  # Limitar os valores negativos
    
    if kplot == freqplot:
        kplot = 0
        XX = str(tempo)
        
        # Salvar resultados
        np.savetxt(os.path.join(output_dir, f"fren_exp{XX}.dat"), fren)
        
        # Plotagem
        plt.figure(1)
        plt.plot(xgrid, fcin, 'r', linewidth=2, label='Condio inicial')
        plt.plot(xgrid, fren, linewidth=2, label='Soluo no tempo')
        plt.grid(True)
        plt.axis([xgrid[0], xgrid[-1], 0, 100])
        plt.title(f'Difuso - Explcito - Tempo {tempo / 3600:.2f} horas', fontsize=12)
        plt.xlabel('Distncia na Grade (m)', fontsize=12)
        plt.ylabel('Concentrao', fontsize=12)
        plt.legend()
        
        # Salvar grfico
        plt.savefig(os.path.join(output_dir, f"tempo_exp{XX}.jpg"))
        plt.pause(0.5)
        plt.clf()  # Limpar figura aps plotagem

    # Atualizar a soluo
    fatu = fren.copy()

plt.show()
```
