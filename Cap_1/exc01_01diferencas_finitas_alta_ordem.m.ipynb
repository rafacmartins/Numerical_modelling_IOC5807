```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros
dx = 0.1
x = np.arange(0, np.pi + dx, dx)
f = np.sin(x)
deranalit = np.cos(x)

jmax = len(x)
dfava = np.zeros(jmax)
dfret = np.zeros(jmax)
dfcen = np.zeros(jmax)

# Derivadas por diferenças finitas de alta ordem
dfava[:jmax-2] = (-3 * f[:jmax-2] + 4 * f[1:jmax-1] - f[2:jmax]) / (2 * dx)
dfret[2:jmax] = (3 * f[2:jmax] - 4 * f[1:jmax-1] + f[:jmax-2]) / (2 * dx)
dfcen[2:jmax-2] = 2 * (f[3:jmax-1] - f[1:jmax-3]) / (2 * dx) - (f[4:jmax] - f[:jmax-4]) / (4 * dx)

# Gráfico da função e das derivadas
plt.figure(1)
plt.plot(x, f, label="Função", linewidth=2)
plt.plot(x[:jmax-2], dfava[:jmax-2], 'r', label="Avançada", linewidth=2)
plt.plot(x[2:jmax], dfret[2:jmax], 'g', label="Retardada", linewidth=2)
plt.plot(x[2:jmax-2], dfcen[2:jmax-2], 'k', label="Centrada", linewidth=2)
plt.title('Função (azul) e Dif. Finitas: Av (vermelho), Ret (verde), Centr (preto)', fontsize=12)
plt.xlabel('x', fontsize=12)
plt.ylabel('f, df/dx alta ordem', fontsize=12)
plt.grid(True)
plt.legend()
plt.show()

# Cálculo dos erros
difer1 = np.zeros(jmax)
difer2 = np.zeros(jmax)
difer3 = np.zeros(jmax)

difer1[:jmax-2] = deranalit[:jmax-2] - dfava[:jmax-2]
difer2[2:jmax] = deranalit[2:jmax] - dfret[2:jmax]
difer3[2:jmax-2] = deranalit[2:jmax-2] - dfcen[2:jmax-2]

# Gráfico dos erros
plt.figure(2)
plt.plot(x[:jmax-2], difer1[:jmax-2], 'r', label="Erro Avançada", linewidth=2)
plt.plot(x[2:jmax], difer2[2:jmax], 'g', label="Erro Retardada", linewidth=2)
plt.plot(x[2:jmax-2], difer3[2:jmax-2], 'k', label="Erro Centrada", linewidth=2)
plt.title('Erros p/ Deriv. Analítica: Av (vermelho), Ret (verde), Centr (preto)', fontsize=12)
plt.xlabel('x', fontsize=12)
plt.ylabel('Erro - alta ordem', fontsize=12)
plt.grid(True)
plt.legend()
plt.axis([x[0], x[-1], -0.004, 0.004])
plt.show()

# Estatísticas dos erros
estat_avanc = [np.mean(difer1), np.median(difer1), np.std(difer1), np.min(difer1), np.max(difer1)]
estat_retar = [np.mean(difer2), np.median(difer2), np.std(difer2), np.min(difer2), np.max(difer2)]
estat_centr = [np.mean(difer3), np.median(difer3), np.std(difer3), np.min(difer3), np.max(difer3)]

print("Estatísticas - Diferença Avançada:", estat_avanc)
print("Estatísticas - Diferença Retardada:", estat_retar)
print("Estatísticas - Diferença Centrada:", estat_centr)
```
