```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros
dx = 0.1
x = np.arange(0, np.pi + dx, dx)
f = x * np.sin(x) * np.exp(-x**2)

jmax = len(x)
dfava1 = np.zeros(jmax)
dfret1 = np.zeros(jmax)
dfcen2 = np.zeros(jmax)
dfava2 = np.zeros(jmax)
dfret2 = np.zeros(jmax)
dfcen3 = np.zeros(jmax)

# Derivadas por diferenças finitas de baixa ordem
dfava1[:jmax-1] = (f[1:jmax] - f[:jmax-1]) / dx
dfret1[1:jmax] = (f[1:jmax] - f[:jmax-1]) / dx
dfcen2[1:jmax-1] = (f[2:jmax] - f[:jmax-2]) / (2 * dx)

# Gráfico das derivadas de baixa ordem
plt.figure(1)
plt.plot(x, f, label="Função", linewidth=2)
plt.plot(x[:jmax-1], dfava1[:jmax-1], 'r', label="Avançada", linewidth=2)
plt.plot(x[1:jmax], dfret1[1:jmax], 'g', label="Retardada", linewidth=2)
plt.plot(x[1:jmax-1], dfcen2[1:jmax-1], 'k', label="Centrada", linewidth=2)
plt.title('Função (azul) e Dif. Finitas: Av (vermelho), Ret (verde), Centr (preto)', fontsize=12)
plt.xlabel('x', fontsize=12)
plt.ylabel('f, df/dx - baixa ordem', fontsize=12)
plt.grid(True)
plt.legend()
plt.show()

# Derivadas por diferenças finitas de alta ordem
dfava2[:jmax-2] = (-3 * f[:jmax-2] + 4 * f[1:jmax-1] - f[2:jmax]) / (2 * dx)
dfret2[2:jmax] = (3 * f[2:jmax] - 4 * f[1:jmax-1] + f[:jmax-2]) / (2 * dx)
dfcen3[2:jmax-2] = 2 * (f[3:jmax-1] - f[1:jmax-3]) / (2 * dx) - (f[4:jmax] - f[:jmax-4]) / (4 * dx)

# Gráfico das derivadas de alta ordem
plt.figure(2)
plt.plot(x, f, label="Função", linewidth=2)
plt.plot(x[:jmax-2], dfava2[:jmax-2], 'r', label="Avançada", linewidth=2)
plt.plot(x[2:jmax], dfret2[2:jmax], 'g', label="Retardada", linewidth=2)
plt.plot(x[2:jmax-2], dfcen3[2:jmax-2], 'k', label="Centrada", linewidth=2)
plt.title('Função (azul) e Dif. Finitas: Av (vermelho), Ret (verde), Centr (preto)', fontsize=12)
plt.xlabel('x', fontsize=12)
plt.ylabel('f, df/dx - alta ordem', fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```
