```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Criar diretório de saída se não existir
output_dir = "result"
os.makedirs(output_dir, exist_ok=True)

# Parâmetros iniciais
jmax = 500   # número de pontos de grade
nmax = 3800  # número de passos de tempo

# Condição inicial
fatu = np.zeros(jmax)
fatu[224:275] = 99  # Definir intervalo de j para sinal retangular (ajuste de índice)
fcin = fatu.copy()

d = 10     # Difusividade em m²/s
dt = 2     # Passo de tempo em s
dx = 8     # Espaçamento da grade em m
q = d * dt / (dx * dx)
xgrid = np.arange(jmax) * dx

# Esquema explícito avançado no tempo e centrado no espaço
fren = fatu.copy()
freqplot = 360
kplot = 359

# Loop temporal
for n in range(2, nmax + 1):
    tempo = n * dt
    kplot += 1
    
    # Fórmula explícita para a equação de difusão
    fren[1:jmax-1] = fatu[1:jmax-1] + q * (fatu[2:jmax] - 2 * fatu[1:jmax-1] + fatu[0:jmax-2])
    fren[fren < 0] = 0  # Limitar os valores negativos
    
    if kplot == freqplot:
        kplot = 0
        XX = str(tempo)
        
        # Salvar resultados
        np.savetxt(os.path.join(output_dir, f"fren_exp{XX}.dat"), fren)
        
        # Plotagem
        plt.figure(1)
        plt.plot(xgrid, fcin, 'r', linewidth=2, label='Condição inicial')
        plt.plot(xgrid, fren, linewidth=2, label='Solução no tempo')
        plt.grid(True)
        plt.axis([xgrid[0], xgrid[-1], 0, 100])
        plt.title(f'Difusão - Explícito - Tempo {tempo / 3600:.2f} horas', fontsize=12)
        plt.xlabel('Distância na Grade (m)', fontsize=12)
        plt.ylabel('Concentração', fontsize=12)
        plt.legend()
        
        # Salvar gráfico
        plt.savefig(os.path.join(output_dir, f"tempo_exp{XX}.jpg"))
        plt.pause(0.5)
        plt.clf()  # Limpar figura após plotagem

    # Atualizar a solução
    fatu = fren.copy()

plt.show()
```
